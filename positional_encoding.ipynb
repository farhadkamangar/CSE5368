```python
# Based on the sinusoidal positional encoding from "Attention Is All You Need"

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import ipywidgets as widgets

def positional_encoding(max_len, d_model):
    """
    Create positional encoding matrix of shape [max_len, d_model].

    PE(pos, 2i)   = sin(pos / 10000^(2i/d_model))
    PE(pos, 2i+1) = cos(pos / 10000^(2i/d_model))
    """
    pe = np.zeros((max_len, d_model), dtype=np.float64)
    positions = np.arange(max_len).reshape(-1, 1)              # [max_len, 1]
    div_term = np.exp(np.arange(0, d_model, 2) * (-np.log(10000.0) / d_model))

    pe[:, 0::2] = np.sin(positions * div_term)                 # even dimensions
    pe[:, 1::2] = np.cos(positions * div_term[:pe[:, 1::2].shape[1]])  # odd dimensions

    return pe

def show_positional_encoding(d_model=16, position=0, max_len=50):
    # Make sure position stays valid
    position = min(position, max_len - 1)

    pe = positional_encoding(max_len=max_len, d_model=d_model)
    vec = pe[position]

    print(f"Positional encoding matrix shape: {pe.shape}")
    print(f"Selected position: {position}")
    print(f"Embedding dimension (d_model): {d_model}")
    print("\nEncoding vector at selected position:")
    print(np.round(vec, 6))

    # Plot 1: heatmap of all positions and dimensions
    plt.figure(figsize=(10, 5))
    plt.imshow(pe, aspect='auto')
    plt.colorbar()
    plt.xlabel("Embedding dimension")
    plt.ylabel("Position")
    plt.title(f"Sinusoidal Positional Encoding Heatmap (max_len={max_len}, d_model={d_model})")
    plt.show()

    # Plot 2: vector values for one selected position
    plt.figure(figsize=(10, 4))
    plt.plot(np.arange(d_model), vec, marker='o')
    plt.xlabel("Dimension index")
    plt.ylabel("Encoding value")
    plt.title(f"Positional Encoding Vector at Position {position}")
    plt.grid(True)
    plt.show()

# Interactive sliders
interact(
    show_positional_encoding,
    d_model=IntSlider(value=16, min=2, max=128, step=2, description='d_model'),
    position=IntSlider(value=0, min=0, max=49, step=1, description='position'),
    max_len=IntSlider(value=50, min=5, max=200, step=5, description='max_len')
);
```
