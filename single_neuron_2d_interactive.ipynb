#  Single Neuron and Activation Functions

This example shows how to calculate the the output of a single neuron using different activation functions.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Optional: slightly crisper plots
plt.rcParams['figure.dpi'] = 120
```

```python
# Define different activation functions
def identity(net):
    return net

def hardlimit(net):
  return np.where(net >= 0, 1, 0)
def sigmoid(net):
    return 1 / (1 + np.exp(-net))

def tanh(net):
    return np.tanh(net)

def relu(net):
    return np.maximum(0, net)

def leaky_relu(net, alpha=0.1):
    return np.where(net >= 0, net, alpha*net)

ACTS = {
    "identity": identity,
    "hardlimit": hardlimit,
    "sigmoid": sigmoid,
    "tanh": tanh,
    "ReLU": relu,
    "leaky ReLU": leaky_relu
}
```

```python
def neuron_demo(
    x1=0.5, x2=0.5, x3=0.5,
    w1=1.0, w2=1.0, w3=1.0,
    b=0.0,
    activation="sigmoid",
    alpha=0.1
):
    # Forward pass
    x = np.array([x1, x2, x3], dtype=float)
    w = np.array([w1, w2, w3], dtype=float)
    net = float(np.dot(x,w) + b)

    f = ACTS[activation]
    if activation == "leaky ReLU":
        a = float(f(net, alpha=alpha))
    else:
        a = float(f(net))



    # Plot activation curve and current point
    zmin, zmax = -10, 10
    zs = np.linspace(zmin, zmax, 1000)

    if activation == "leaky ReLU":
        ys = f(zs, alpha=alpha)
    else:
        ys = f(zs)

    plt.figure(figsize=(8, 4))
    plt.plot(zs, ys)
    plt.axvline(0, linestyle="--")
    plt.scatter([net], [a], s=80)
    plt.title(f"Activation: {activation}   (net={net:.3f} , a={a:.3f})")
    plt.xlabel("net")
    plt.ylabel("a = f(net)")
    plt.grid(True)
    plt.show()
# Printing
    print("Single Neuron")
    print("-" * 40)
    print(f"x = [{x1:.3f}, {x2:.3f}, {x3:.3f}]")
    print(f"w = [{w1:.3f}, {w2:.3f}, {w3:.3f}]")
    print(f"b = {b:.3f}")
    print(f"net = x.w + b = {net:.6f}")
    print(f"a = f(net) with {activation} = {a:.6f}")
```

```python
# --- Controls (sliders) ---
x1 = widgets.FloatSlider(value=0.5, min=-2, max=2, step=0.01, description='x1', continuous_update=True)
x2 = widgets.FloatSlider(value=0.5, min=-2, max=2, step=0.01, description='x2', continuous_update=True)
x3 = widgets.FloatSlider(value=0.5, min=-2, max=2, step=0.01, description='x3', continuous_update=True)
w1 = widgets.FloatSlider(value=1.0, min=-5, max=5, step=0.01, description='w1', continuous_update=True)
w2 = widgets.FloatSlider(value=1.0, min=-5, max=5, step=0.01, description='w2', continuous_update=True)
w3 = widgets.FloatSlider(value=1.0, min=-5, max=5, step=0.01, description='w3', continuous_update=True)
b  = widgets.FloatSlider(value=0.0, min=-5, max=5, step=0.01, description='b', continuous_update=True)
activation = widgets.Dropdown(options=list(ACTS.keys()), value='identity', description='activation:')
alpha = widgets.FloatSlider(value=0.1, min=0.0, max=1.0, step=0.01, description='Leaky α', continuous_update=True)
# Group sliders vertically (left)
controls = widgets.VBox(
    [x1, x2, x3, w1, w2, w3, b, activation, alpha],
    layout=widgets.Layout(width='320px', margin='0 20px 0 0')
)
# Output area (right)
out = widgets.interactive_output(
    neuron_demo,
    {
        'x1': x1, 'x2': x2, 'x3': x3,
        'w1': w1, 'w2': w2, 'w3': w3,
        'b': b,
        'activation': activation,
        'alpha': alpha,
    }
)
# Side-by-side layout
display(
    widgets.HBox(
        [controls, out],
        layout=widgets.Layout(align_items='flex-start')
    )
)
```
